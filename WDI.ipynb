## 출생 시 평균 수명(년) & 1인당 GDP(현재 US$)

```python
df[df["Indicator Name"]=="Life expectancy at birth, total (years)"].head(2)
```

```python
df[df["Indicator Name"]=="GDP per capita (current US$)"].head(2)
```

```python
df_lg = df[(df["Indicator Name"]=="Life expectancy at birth, total (years)") | (df["Indicator Name"]=="GDP per capita (current US$)")].copy()
df_lg.head()
```

```python
df_lg.shape
```

```python
df_lg = df_lg.drop(["Country Code", "Indicator Code"], axis=1)
df_lg.head(2)
```

- unnamed column 삭제 : <https://stackoverflow.com/questions/43983622/remove-unnamed-columns-in-pandas-dataframe>

```python
df_lg.drop(df_lg.columns[df_lg.columns.str.contains('unnamed',case = False)], axis = 1, inplace = True)
df_lg.head(2)
```

- 데이터 재구조화 : <https://diane-space.tistory.com/103>

```python
df_lg.pivot(index="Country Name", columns="Indicator Name", values=["2017", "2018", "2019"]).head()
```

```python
df_lg.unstack()
```

```python
df_lg.pivot_table(index=["Indicator Name"])
```

```python
df_lg.groupby("Indicator Name").describe()
```

```python
df_lg_r = pd.melt(df_lg, id_vars=["Country Name", "Indicator Name"])
df_lg_r.head(2)
```

```python
df_lg_r = df_lg_r.rename(columns={"variable": "Year"})
df_lg_r.head(2)
```

```python
df_lg_rr = df_lg_r.pivot(index=["Country Name", "Year"], columns="Indicator Name", values="value")
df_lg_rr.head(2)
```

```python
df_lg_rr = df_lg_rr.rename(columns={"GDP per capita (current US$)": "GDP", "Life expectancy at birth, total (years)": "Life"})
df_lg_rr.head(2)
```
